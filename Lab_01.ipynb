**Exemplo 01** - Expansão da função $f(x) = e^x$, via Série de MacLaurin

$$
f(x) = e^x \approx \sum_{k=0}^{N}\frac{x^k}{k!}
$$

```python
import numpy as np
import matplotlib.pyplot as plt
import math

def mac(x, N):
    return sum((x**k) / math.factorial(k) for k in range(N+1))

x = np.linspace(0,2*np.pi,100)
result = mac(x, 10)
print(f"Resultado {result}")
plt.plot(result)
plt.grid(True)
plt.show()
```

**Exemplo 03** : Use a expansão em Série de MacLaurin para mostrar que

\begin{equation}
e^{ix} = cos\,x + i\,sen\,x,\,\,\,\,\,\,i = \sqrt{-1}.
\end{equation}

```python
import numpy as np
import matplotlib.pyplot as plt
import math

def mac(x, N):
    return sum((x**k) / math.factorial(k) for k in range(N+1))

x= np.linspace(0,2*np.pi,100)

result = mac(x, 20)
result = result**1j
result2 = np.cos(x) + 1j*np.sin(x)
print(f"Resultado1 {result}")
print(f"Resultado2 {result2}")
plt.plot(result)
plt.plot(np.imag(result))
plt.grid(True)
plt.show()
```

**03** - Escreva a expansão em série de Taylor para $f(x) = e^{x^2}$ em torno da origem. Defina uma função $exp$\_$dupla(x,n),$ para calcular uma aproximação para $f(x)$ utilizando os $n$ primeiros termos da expansão. Adapte a função para receber como parâmetro de entrada um vetor.



```python
import numpy as np
import matplotlib.pyplot as plt
import math

def exp_dupla(x, n):
    soma = np.zeros_like(x, dtype= float)
    soma += sum(np.pow(x**2,k) / math.factorial(k) for k in range(n+1))
    return soma

x = np.linspace(0.1,0.5,100)
n = 2
resultado = exp_dupla(x, n)
plt.plot(resultado)
plt.grid(True)
plt.show()
```
