**Exemplo 01** - Expansão da função $f(x) = e^x$, via Série de MacLaurin

$$
f(x) = e^x \approx \sum_{k=0}^{N}\frac{x^k}{k!}
$$

```python
import numpy as np
import matplotlib.pyplot as plt
import math


def mac(x: complex, N: int = 0) -> list[complex]:
    soma = 0
    valores: list[complex] = []
    for k in range(N+1):
        soma += (x**k) / math.factorial(k)
        valores.append(soma)
    return valores


result = mac(1, 10)
print(f"Resultado {result[-1]}")
plt.plot(result)
plt.show()
```

**Exemplo 02** - Implementar a função

\begin{equation}
     f(x) = sen(x) \approx \sum_{k=0}^{N} (-1)^k \frac{x^{2k+1}}{(2k+1)!}
\end{equation}

```python
def func(x: int, N: int = 0) -> list[float]:
    valores: list[float] = []
    soma = 0
    for k in range(N+1):
        soma += ((-1)**k) * ((x**((2*k)+1)) / math.factorial((2*k) + 1))
        valores.append(soma)

    return valores


result = func(1, 10)
print(f"Resultado {result[-1]}")
plt.plot(result)
plt.show()
```
